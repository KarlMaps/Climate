```python
import os
import matplotlib.pyplot as plt
import numpy as np
import rasterio
import rasterio.coords as CRS
import rioxarray as rxr
import seaborn as sns
from rasterio.plot import show_hist
```

```python
slr=rasterio.open('/Users/olivierlauret/DEV/DATA/CMIP6_SLR_2021-2040_SSP5-8.5_Annual.tiff')
slr2=rasterio.open('/Users/olivierlauret/DEV/DATA/CMIP6_SLR_2081-2100_SSP5-8.5 _Annual.tiff')
```

```python
%matplotlib widget
fig, axhist = plt.subplots(1, 1)
show_hist(slr, bins=100, histtype='stepfilled',
          lw=0.0, stacked=False, alpha=0.5, label="RCP 4.5",ax=axhist)
show_hist(slr2, bins=100, histtype='stepfilled',
          lw=0.0, stacked=False, alpha=0.6, label="RCP 8.5", ax=axhist)
axhist.set_xlabel('Sea Level')
axhist.set_ylabel('Frequency')
axhist.set_title('2021-2040')
sns.set_theme(style="dark", palette='vlag')
```

```python
array = slr2.read(1)
```

```python
array
```

```python
fig, axhist = plt.subplots(1, 1,figsize=[11,5])
plt.rcParams["figure.autolayout"] = True
cm=sns.histplot(data=array, kde='true', stat='percent', bins=80)
#axhist.axes.get_xaxis().set_visible(False)
plt.legend(loc='lower center')
plt.show()
slr2.close()
```

```python
fig, axhist = plt.subplots(1, 1,figsize=[15,8])
plt.rcParams["figure.autolayout"] = True
cm=sns.kdeplot(data=array, fill=True, vertical = True, shade=True, palette="crest",alpha=.5, linewidth=0)
#axhist.axes.get_xaxis().set_visible(False)
plt.legend(loc='lower center')
plt.show()
slr2.close()
```
